```python
class Barang:
  def __init__(self,kode,nama,harga,stok):
    self.kode=kode
    self.nama=nama
    self.harga=harga
    self.stok=stok


  def update_stok(self, jumlah):
    self.stok += jumlah

  def info_barang(self):
    return f"Kode\t: {self.kode}\nNama\t: {self.nama}\nHarga\t: {self.harga}\nStok\t: {self.stok}"



pensil=Barang('AAO','Pensil',1000,100)
print(pensil.info_barang())
pensil.update_stok(-20)
print("Stok pensil saat ini : ", pensil.stok)
```
